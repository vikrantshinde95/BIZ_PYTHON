```python
### generate for loop for annotate

friends = [70,65,72,63,44,52,78,55]
minutes = [140,100,80,69,125,90,70,150]
labels = ["a","b","c","d","e","f","g","h"]

plt.scatter(friends,minutes,marker="*", s=100)

for friend_count ,minute_count, label in zip(friends,minutes,labels):  # squence of variable should match with what we pass in zip

    plt.annotate(label,xy=(friend_count,minute_count),xytext = (4,5), textcoords="offset points")
    
plt.title("daily minutes vs number of friends")
plt.xlabel("# of friends")
plt.ylabel("# Daily minutes spent on calls")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
slices = [7,12,9,14]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow= True,
        textprops = {'fontsize':14, 'c':'g'},
        explode=(0,0,0,0),
        autopct='%1.2f%%')
plt.title('Day wise activity time spent - pattern')
```
