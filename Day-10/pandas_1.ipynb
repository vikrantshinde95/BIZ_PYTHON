### bulid datetimeindex with pd.date_range() method

```python
idx = pd.date_range("jan 07,2024", periods=7, freq="D")
idx
```

```python
idx = pd.date_range("01/07/2024", periods=7, freq="D")  # D- daily
idx
```

```python
idx = pd.date_range("01/07/2024", periods=7, freq="W-SUN")  # W- Weekly
idx
```

```python
idx = pd.date_range("01-02-2025", periods=7, freq="D")  # W- Weekly
idx
```

```python
pd.to_datetime(["Jan 01,2019","01/02/2019","03-Jan-2019",None])
```

```python
pd.to_datetime(["Jan 01,2019","01/02/2019","01-Jan-2019",None])
```

### to convert any text string into datetime format

```python
pd.to_datetime(["Jan 01, 2052","01-01-2019","02/01/2019","03-Jan-2018",None],format="mixed")
```

```python
dates = np.array(["2015-07-10","2015-07-15","2015-07-20","2025-07-25"], dtype = "datetime64[D]")
dates
```

```python
# convert to an index

idx = pd.DatetimeIndex(dates)
idx
```

```python
datetime_df = pd.DataFrame(idx)
datetime_df
```
